```python
# This notebook checks whether our Python environment is set correctly.
# If this works, everything else will work.

import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn

print("Python version:", sys.version)
print("NumPy version:", np.__version__)
print("Pandas version:", pd.__version__)
print("Scikit-learn version:", sklearn.__version__)

# Simple graph to test plotting
x = np.arange(0, 10)
y = x ** 2

plt.plot(x, y)
plt.title("Environment Test Plot")
plt.xlabel("x")
plt.ylabel("x squared")
plt.show()
```
